git_conn の　trainのデータより、train_m2（m2 は　マークⅡの意）を作る。
train_m2 は、trainの全要素で構成され、ラベル y_train_m2 をもつ。

```python
import os
import shutil
M_dic = {'ARASHI': 0,'MRCHILDREN': 1,'NAGABUCHI': 2,'SOUTHEN': 3,'SPITZ': 4,'TANIMURA': 5,'YONE': 6,'YUZU': 7}

cnt = 1
y_train_m2 = []
work_dir = "train"
to_dir = "train_m2"
for current_dir,dirs,files in os.walk(work_dir):
    for file in files:
        # pngデータのみ以下の処理を行う
        if file.rsplit(".")[-1] == "png":
            newfilename = f'tr_{cnt:03}.png'
            shutil.copy(os.path.join(current_dir,file),os.path.join(to_dir,newfilename))
            label_name = current_dir.rsplit("/")[-1]
            y_train_m2.append(M_dic[label_name])
            cnt += 1
```

```python
print(y_train_m2)
```

```python
import pandas as pd
df = pd.DataFrame(y_train_m2)
df.to_csv("y_train_m2.csv",index=False,header=False)
```
